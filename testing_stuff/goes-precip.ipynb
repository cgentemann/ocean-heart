```python
from goes2go import GOES
import warnings
warnings.filterwarnings('ignore')
```

```python
# ABI Multi-Channel Cloud Moisture Imagry Product
G = GOES(satellite=16, product="ABI-L2-RRQPEF", domain='F')
```

```python
# Download and read the data as an xarray Dataset nearest a specific time
ds = G.nearesttime('2024-06-22')
```

```python
ds
```

```python
import numpy as np
tem = np.log(ds.RRQPE)
tem.plot(vmin=0,vmax=2)
```

```python
import matplotlib.pyplot as plt
ds = GOES().latest()
ax = plt.subplot(projection=ds.rgb.crs)
ax.imshow(ds.rgb.TrueColor(), **ds.rgb.imshow_kwargs)
ax.coastlines()
```
